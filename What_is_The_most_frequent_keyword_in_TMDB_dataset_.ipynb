In this notebook we will going to answer a few research analytical qustions about Movies production industry using TMDB dataset

Research Question 1 :-What is The most frequent keyword in TMDB dataset?

import some packages

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

load data and cleaning it 

```python
path="/content/tmdb-movies.csv"
df=pd.read_csv(path)
df_keywords=df["keywords"].dropna(axis=0) #remove nan values of keywords column
print("the # null values:",df_keywords.isnull().sum())#check for null values

df_keywords.head()
```

Count the words to determine the most frequent word

```python

keywords={}#empty dict
for i in range (df_keywords.size):
  try:
    KeyWords_list=df_keywords[i].split(sep="|")#split the record to extract the words
    for j in (KeyWords_list):
        if j in keywords:
            keywords[j]+=1
        else:
            keywords[j]=1
  except:
      pass
#sort the dict in descending order
keywords_sorted={k: v for k, v in sorted(keywords.items(), key=lambda item: item[1],reverse = True)}
keywords_sorted=list(keywords_sorted.items())[:50] #select the first 50 item to plot them




```

Plotting The most frequent word

```python
plt.figure(figsize=(20, 10))
for i in range (len(keywords_sorted)):
  rects1=plt.bar(keywords_sorted[i][0],keywords_sorted[i][1],0.4)
plt.xticks(rotation=85, fontsize=15)
plt.xlabel("Keywords")
plt.ylabel('Number of Samples')
plt.title("Answer to our Research Question 1\n What is The most frequent keyword in TMDB dataset? ")
plt.show()
```
